```python
from urllib.request import urlopen
```

```python
ur1 = 'https://en.wikipedia.org/wiki/Website'
```

```python
P1 = urlopen(ur1)
```

```python
P1.status
```

```python
P2 = P1.read()
```

```python
P2
```

```python
P3 = P2.decode('utf-8')
```

```python
print(P3)
```
